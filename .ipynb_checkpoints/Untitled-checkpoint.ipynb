```python
threshold = 0.5 * len(df) 
```

```python
df_cleaned = df.dropna(axis =1 , thresh = threshold)
```

```python
df_cleaned
```

```python
numeric_cols = df_cleaned.select_dtypes(include=['number'])
```

```python
mean_zero = numeric_cols.mean() == 0
var_zero = numeric_cols.var() == 0
```

```python
cols_to_remove = numeric_cols.columns[mean_zero & var_zero]
```

```python
df_cleaned = df_cleaned.drop(columns=cols_to_remove)
```

```python
df_cleaned
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Aggregate the data: count occurrences of each value
bankruptcy_counts = df_cleaned['pub_rec_bankruptcies'].value_counts().sort_index()

# Create a line chart
plt.figure(figsize=(10, 6))
plt.plot(bankruptcy_counts.index, bankruptcy_counts.values, marker='o', linestyle='-', color='blue')

# Add labels and title
plt.title("Distribution of Public Record Bankruptcies", fontsize=16)
plt.xlabel("Number of Bankruptcies", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.grid(alpha=0.3)

# Show the plot
plt.show()
```
